```python
import pandas as pd 
import numpy as np
import os
import matplotlib.pyplot as plt
from collections import deque
import random

from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder as encoder
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization, Flatten
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import TensorBoard, ModelCheckpoint

np.warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 10)
cwd = os.path.dirname(os.getcwd())
```

```python
def preprocess_train_df( df, SEQUENCE_LEN,  ):
    df_aux = df
    
    #------------- scaling --------------#
    df_aux = df_aux.drop('date', axis=1)
    for col in df_aux.columns:
        if col in ['close' , 'open' ,'low' , 'high']:
            df_aux[col] = df_aux[col].pct_change()
            df_aux.dropna(inplace=True)
            df_aux[col] = preprocessing.scale(df_aux[col].values)
    df_aux.dropna(inplace=True)
    
    #--------------- create seqs -------------#
    sequential_data = []
    prev_days = deque(maxlen=SEQUENCE_LEN)
    for i in df_aux.values:
        prev_days.append( [n for n in i[:-1]] )
        if len(prev_days) == SEQUENCE_LEN:
            sequential_data.append( [ np.array( prev_days ), i[-1] ] )

    #------------------------------------------#
    random.shuffle(sequential_data)
    downs = []
    ups = []

    for seq, target in sequential_data:
        if target == 0:
            downs.append([seq,target])
        elif target == 1:
            ups.append([seq,target])

    random.shuffle(downs)
    random.shuffle(ups)

    # SCALE NUMBER OF EVENTS
    lower = min(len(downs),len(ups))
    ups = ups[:lower]
    downs = downs[:lower]

    sequential_data = ups+downs
    random.shuffle(sequential_data)

    X, y = [] , []

    for seq,target in sequential_data:
        X.append(seq)
        y.append(target)

    return np.array(X), np.array(y)

def preprocess_test_df( df, SEQUENCE_LEN ):
    #scaling
    df_aux = df
    df_aux = df_aux.drop('date', axis=1)
    
    for col in df_aux.columns:
        if col in ['close' , 'open' ,'low' , 'high']:
            df_aux[col] = df_aux[col].pct_change()
            df_aux.dropna(inplace=True)
            df_aux[col] = preprocessing.scale(df_aux[col].values)
    df_aux.dropna(inplace=True)
    
    sequential_data = []
    prev_days = deque(maxlen=SEQUENCE_LEN)
    for i in df_aux.values:
        prev_days.append( [n for n in i[:-1]] )
        if len(prev_days) == SEQUENCE_LEN:
            sequential_data.append( [ np.array( prev_days ), i[-1] ] )

    X,y = [] , []
    for seq,target in sequential_data:
        X.append(seq)
        y.append(target)

    return np.array(X), np.array(y)
```

```python
#----------- MODEL VARS --------------------------#
FUTURE_PERIOD = 1
TRAIN_TEST_SPLIT = 0.2 
SEQUENCE_LEN = 5
EPOCHS = 20
BATCH_SIZE = 100
```

```python
#---------- READ CSV -------------#
df = pd.read_csv(cwd+'/data/BTCUSDT_2021-07-30_2021-08-07_1m.csv', index_col = 0)
df.columns = [ col.lower() for col in df.columns ]
df = df[['date','open', 'high', 'low', 'close']]
display(df.head())
#---------- CREATE LABEL ---------#
df['close_target'] = df['close'].shift(-FUTURE_PERIOD)
df['label'] =  df.apply(lambda row: 1 if row['close_target']>=row['close'] else 0, axis = 1)
df = df.drop('close_target', axis = 1)
display(df.head())
```

```python
#--------- TRAIN/TEST SPLIT -----------#
times = sorted(df.index.values)
times = times[-int(TRAIN_TEST_SPLIT*len(times))]

train_df = df[(df.index < times)]
val_df = df[(df.index >= times)]
test_df = val_df.copy()

X_train, y_train = preprocess_train_df(train_df, SEQUENCE_LEN)
X_val, y_val = preprocess_test_df(val_df, SEQUENCE_LEN)
```

```python
# DNN MODEL

model = Sequential()

model.add(LSTM(128, input_shape = (X_train.shape[1:]), return_sequences=True  ))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(LSTM(32, input_shape = (X_train.shape[1:])))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(100, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(30, activation = 'relu'))

model.add(Dense(1, activation = 'sigmoid'))
```

```python
#OPTIMIZER
opt = tf.keras.optimizers.Adam(lr = 1e-3, decay = 1e-6)
model.compile(loss = 'binary_crossentropy', optimizer = opt, metrics = ['accuracy'])

#FIT 
history = model.fit(X_train, y_train, batch_size = BATCH_SIZE, epochs = 300, validation_data = (X_val, y_val), verbose= 1)
```
